## 열 별 그룹화

```python
titanic.groupby('sex').count() #그룹바이에 적힌(sex)에따라서 그룹화하여 각각의 빈도수 전체를 구해서 보여줌
```

```python
titanic.groupby('who').count() #index(행)이 who라는 의미
```

```python
titanic.groupby('embark_town').count() #embarktown그냥 경유지임
```

```python
titanic.groupby('pclass')["fare"].mean() #pclass(등급)별로 좌석을 그룹화해서 요금의 평균을 구해라
```
