##데이터분석을 위한 seaborn 내장(토이) 데이터, titanic

```python
import seaborn as sns
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
titanic.info()
```

```python
sns.countplot(titanic, x = "sex")
```

```python
sns.countplot(titanic, x = "class");
```

```python
sns.countplot(titanic, x = "deck", hue="deck")
```

```python
sns.countplot(titanic, x = "class", hue="alive") #hue는 x축의 first second third를 나눠서 그려라는 의미 3등급이 비율적으로 많이죽었다
```

```python
sns.countplot(titanic, x = "who", hue="alive", stat = "percent");
```

## 열 별 그룹화

```python
titanic.groupby('sex').count() #그룹바이에 적힌(sex)에따라서 그룹화하여 각각의 빈도수 전체를 구해서 보여줌
```

```python
titanic.groupby('who').count() #index(행)이 who라는 의미
```

```python
titanic.groupby('embark_town').count() #embarktown그냥 경유지임
```

```python
titanic.groupby('sex').count()
```
